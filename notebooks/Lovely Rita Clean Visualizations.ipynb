```python
zip_viol = zip_viz_data[['Ticket_Issue_Date','Zipcode','Violation_External_Code']]
grouped_zip = zip_viol.set_index('Ticket_Issue_Date').groupby([pd.TimeGrouper('A'),'Zipcode']).count()
grouped_zip.head(20)
```

```python
bar = grouped_zip.plot(kind = 'bar', figsize = (20,6), color = 'Orange', grid = False, \
                            legend = False)
```

```python
zip_viz_data['Ticket_Year'] = zip_viz_data['Ticket_Issue_Date'].dt.year
line_df = (zip_viz_data[['Ticket_Year','Zipcode']].reset_index()
          .groupby(['Ticket_Year','Zipcode'], as_index=False)
          .count()
          # rename isn't strictly necessary here, it's just for readability
          .rename(columns={'index':'ct'})
       )
line_df.head(20)
```

```python
fig, ax = plt.subplots(figsize = (15,10))

# key gives the group name (i.e. category), data gives the actual values
for key, data in line_df.groupby('Zipcode'):
    data.plot(x='Ticket_Year', y='ct', ax=ax, label=key, yticks= range(0,50000,5000),\
              xticks= range(2012,2017), fontsize= 14)
    
ax.legend(bbox_to_anchor=(1, 1), loc = 'upper left',borderaxespad=1, fontsize = 18,\
          markerscale = .4, frameon = True, framealpha = .1, edgecolor = 'black',\
          labelspacing = .7)
```

Hardly any variation in number of citations by zip code

```python
viol_type_df = (zip_viz_data[['Zipcode','Violation_Description_Long']].reset_index()
          .groupby(['Zipcode','Violation_Description_Long'], as_index=False)
          .count()
          # rename isn't strictly necessary here, it's just for readability
          .rename(columns={'index':'ct'})
       )
viol_type_df.head(20)
```

```python
import seaborn as sns
fg = sns.factorplot(data=viol_type_df, y="Violation_Description_Long", x="ct", row="Zipcode",\
               kind="bar", aspect= 2, size=10)
fg.savefig("Violation_By_Zip.pdf")
```
