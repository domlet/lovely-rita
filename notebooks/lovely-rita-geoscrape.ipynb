```python
import requests
import time
from pymongo import MongoClient
import datetime
```

Static Variables

```python
address_file_01 = "../data/addresses_01.txt"
address_file_02 = "../data/addresses_02.txt"

base_url = "https://maps.googleapis.com/maps/api/geocode/json?&address="
```

Functions

```python
def get_set_from_file(file):
    data = None
    with open(file, 'r') as f:
        data = set([line.strip() for line in f])
    return data

def run_api_pull(base_url, query, client, db, coll_map, coll_data, key_extractor, time_sleep=0.1):
    """
    Function for pulling data from an API and storing it in a mongoDB collection.
    """

    # check if query has already been run
    map_check = client[db][coll_map].find({"_id": query}).count()
    if map_check != 0:
        print "Already Mapped!"
        return None
    else:
        # query has not been run before
        print "Pull Started!"
        print "Sending Request."
        r = requests.get(base_url + query)
        time.sleep(time_sleep) # to prevent lockout
        r_json = r.json()
        if not r_json:
            print "No Repsonse From API!"
            return None        
        elif r_json["status"] != 'OK':
            print "Status not OK!"
            return query
        else:
            print "Response OK!"
            uuid = key_extractor(r_json)

            client[db][coll_map].save({"_id": query, "geo_id": uuid})
            print "Mapping Saved."

            r_json["_id"] = uuid
            r_json["timestamp"] = str(datetime.datetime.now())
            client[db][coll_data].save(r_json)
            print "Data Saved."
            return r_json

def extract_google_geocode_key(response):
    return response["results"][0]["place_id"]
```

```python
address_01 = get_set_from_file(address_file_01)
address_02 = get_set_from_file(address_file_02)
print len(address_01), len(address_02)
```

```python
address_01.intersection(address_02)
```

There is no overlap in the two address sets.

Process for pulling batch of addresses from Google Geocode API

```python
client = MongoClient("127.0.0.1", 27017)
count = 0
error_addresses = []
for address in address_01:
    if count >= 1000:
        break
    response = run_api_pull(base_url, address, client, "lovely-rita", "map", "geo", extract_google_geocode_key)
    if response:
        if type(response) is str:
            error_addresses.append(response)
        else:
            count += 1
```
